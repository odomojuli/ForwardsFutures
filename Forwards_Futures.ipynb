FORWARDS

A **forward** contract is an agreement to buy and sell an asset at a given time in the future, at a price pre-specified today. 

There are two dates that are relevant: (a) Contracting date, (b) settlement date. 

The pre-specified price is called the "forward price" $F$. The spot price is denoted $S$.  

Suppose you are given:

IBM current price = $100

Risk free interest rate (1 yr)= 2% 

Expected return on S&P500 = 10% 

IBM beta = 1.5

What is the one year forward price for IBM?


PAYOFF DIAGRAMS

Payoff diagrams depict the cash flows generated by the security at a given date. The payoff of a long position in a forward at price $F$ is a function of the stock price at maturity $S_T$. 

```python
#PAYOFF DIAGRAM
S = linspace(75,125,51)
F = 105
payoff = S-F
plot(S,payoff)
xlabel("S(T)")
ylabel("payoff")
plot(S,0*S)
grid()
show()
```

For a short position, the payoff is $F-S_T$. Forward contract payoffs are linear in $S_T$. 

The cash flow at inception of a forward is zero.

PRICING BY REPLICATION

Replication of the forward is done by holding the spot asset and borrowing or lending cash such that the payoff of this portfolio is identical to that of the forward. 

The simplest replication of a stock forward is done by physically holding the stock at time zero, and carrying it till maturity. In addition a corresponding borrowing is required to fund the stock purchase. The interest on the borrowing is the **cost-of-carry**. The carry formula for the forward is:
\begin{equation}
F = S e^{rT}
\end{equation}

If there are other carry costs and benefits, these may be used to adjust the formula. Define 

\begin{equation}
M = PV(\mbox{Holding costs}) - PV(\mbox{Holding benefits})
\end{equation}

The NPV of the replicating strategy is 

\begin{equation}
S + M = PV(F) = F e^{-rT}
\end{equation}

Hence the forward price is

\begin{equation}
F = (S+M) e^{rT}
\end{equation}


```python
#Example: Stock forward price
S = 350
rf = 0.03
#No holding costs
#What is the three-month forward price?
F = S*exp(0.03*0.25)
print "Forward price =", F
```

### IMPLIED REPO RATE

Assume a dividend paying stock with dividend rate $d$. The forward price is 

\begin{equation}
F = S e^{(r-d)T}
\end{equation}

Re-arranging we get the implied repo rate:

\begin{equation}
r = d + \frac{1}{T} [\ln F - \ln S]
\end{equation}

This is a synthetic borrowing/lending rate.

### HEDGING

Assume a commitment to buy $Q$ units of the spot asset at date $T$. To hedge use $H$ units of the forward/future today.  There are no cash flows today but we care about minimizing the variance of the cash flows at maturity. Net cash outflow at maturity is

\begin{equation}
Q S_T + (F-S_T) H = (Q-H)S_T + FH
\end{equation}

To minimize the variance of the hedged cash flow, 

\begin{equation}
\min_H (Q-H)S_T + FH \quad \Longrightarrow \quad H = Q
\end{equation}

The optimal hedge ratio is 

\begin{equation}
h^* = \frac{H}{Q} = 1
\end{equation}


### MINIMUM VARIANCE HEDGING

Start with the hedged cash flow and re-arrange, noting that $F_T=S_T$.

\begin{eqnarray*}
Q S_T + (F-F_T) H &=& Q S_T -QS + QS + (F-F_T) H\\
&=& Q \Delta S -H  \Delta F + QS\\
&=& Q[\Delta S - h \Delta F] + QS
\end{eqnarray*}

Minimize the variance of this expression, i.e., find $h$ to minimize

\begin{eqnarray*}
V &=& Var(Q[\Delta S - h \Delta F]) \\
&=& Q^2[Var(\Delta S - h \Delta F)]\\
&=& Q^2 [Var(\Delta S) + h^2 Var(\Delta F) + 2h Cov(\Delta S,\Delta F)]
\end{eqnarray*}

Differentiate

\begin{equation}
\frac{dV}{dh} = Q^2 [2h Var(\Delta F) + 2 Cov(\Delta S, \Delta F)] = 0
\end{equation}

The solution is:

\begin{equation}
h^* = \frac{Cov(\Delta S, \Delta F)}{Var(\Delta F)} = \rho \cdot \sqrt{\frac{Var(\Delta S)}{Var(\Delta F)}}
\end{equation}

The minimized variance is
\begin{equation}
Q^2 Var(\Delta S) (1-\rho^2)
\end{equation}

Note that the unhedged variance is simply $Q^2 Var(\Delta S)$. 


Example: $Var(\Delta S)=15$, $Var(\Delta F)=18$, $Corr(\Delta S, \Delta F)=0.96$:

```python
var_ds = 15.
var_df = 18.
corr = 0.96
h_star = corr * sqrt(var_ds/var_df)
print "h_star = ",h_star
```

### REGRESSION FOR HEDGING

```python
from pandas.io.data import DataReader
from datetime import datetime
ge = DataReader("GE",  "yahoo", datetime(2013,1,1), datetime(2015,3,31))
g = ge["Adj Close"]
sp = DataReader("GSPC",  "yahoo", datetime(2013,1,1), datetime(2015,3,31))
s = sp["Adj Close"]
plot(g)
plot(s)
print len(g)
```

```python
import pandas as pd
stkp = pd.DataFrame(vstack((g,s)).T, columns=["GE","SP"])
print stkp.corr()
stkp.head()
```

```python
stkp = stkp.diff(252)
```

```python
cv = array(stkp.cov())
print(cv)
h_star = cv[0,1]/cv[1,1]
print "h_star = ", h_star
```

```python
#REGRESSION

import statsmodels.api as sm
x = stkp["SP"]
y = stkp["GE"]
x = sm.add_constant(x)
res = sm.OLS(y,x).fit()
print res.summary()
```
